```python
def massCrit(z,LWI):
    return ((2.5*10**5)*((1+z)/26)**(-3/2))*(1+6.96(4*np.pi*LWI))

def massA(z):
    return (5.4*10**7)*(((1+z)/11)**(-3/2)) # use whichever is smaller
```

```python
# Using dotted green curve for LW feedback. Using that curve to interpolate LW feedback at each redshift.
# Working under the assumption that the feedback flatlines at redshifts less than 10.
```

```python
def JLW21(z):
    if (z <= 10):
        return 7
    elif ((z > 10) & (z <= 15)):
        return (-1*z)+17
    elif ((z > 15) & (z <= 20)):
        return (-0.32*z)+6.8
    elif ((z > 20) & (z <= 25)):
        return (-0.07*z)+1.8
    elif ((z > 25) & (z <= 30)):
        return (-0.0093*z)+0.2825
    elif ((z > 30) & (z <= 35)):
        return (-0.00069*z)+0.0242
```
