```python
def JLW21(z): #Interpolated Lyman-Werner intensity, as a function of redshift
    if (z <= 10):
        return 7
    elif ((z > 10) & (z <= 15)):
        return (-1*z)+17
    elif ((z > 15) & (z <= 20)):
        return (-0.32*z)+6.8
    elif ((z > 20) & (z <= 25)):
        return (-0.07*z)+1.8
    elif ((z > 25) & (z <= 30)):
        return (-0.0093*z)+0.2825
    elif ((z > 30) & (z <= 35)):
        return (-0.00069*z)+0.0242
```
